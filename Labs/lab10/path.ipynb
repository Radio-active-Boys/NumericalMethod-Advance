```python
import random
import numpy as np
import matplotlib.pyplot as plt

def hill_climbing_function(x, A, B, D, L):
    periodic_component = A * np.sin(B * x)
    exponential_component = np.exp(-x)
    random_component = D * random.uniform(-L, L)
    return periodic_component + exponential_component + random_component

# Example usage:
A = 10    # Amplitude of the periodic component
B = 0.5   # Frequency of the periodic component
D = 2     # Scaling factor for the non-periodic random component
L = 2     # Small limit for the random offset

x_values = np.linspace(0, 10, 1000)
y_values = [hill_climbing_function(x, A, B, D, L) for x in x_values]

plt.plot(x_values, y_values)
plt.title('Hill Climbing Path Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.show()
```
